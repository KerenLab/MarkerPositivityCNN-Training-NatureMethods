# Training

You can also pass the inner `MantisProject…` folder or a flat FOV layout.

Label CSVs: `fov`, `cellID`, `Marker`, `Positive`, … Checkpoints: `<model_name>_<timestamp>_epoch_*.pth` and a final `.pt` under **output** (relative to this repo root).

After updating the main repo, rebuild the bundle with `python scripts/build_training_bundle.py` so this copy stays in sync.
